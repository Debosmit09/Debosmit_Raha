# **TRANSFORMATION**

```python
#Translation
def translate(img2 , x , y):
  transMat = np.float32([[1,0,x],[0,1,y]])
  dimensions = (img.shape[1],img.shape[0])
  return cv2.warpAffine(img,transMat,dimensions)

translated = translate(img2 ,-400 ,-200)
plt.imshow(translated)
```

```python
#Rotation

height,width=img2.shape[:2]
x,y=(width//2,height//2)
m=cv2.getRotationMatrix2D((x,y), 45, 1.0)
rotated = cv2.warpAffine(img2, m, (width, height)) #original image matrix, transformation image
plt.imshow(rotated)
```

```python
#Scaling
stretch_near = cv2.resize(img2, (900, 940), interpolation = cv2.INTER_LINEAR)
plt.imshow(stretch_near)
```

```python
#Shearing
rows, cols = img2.shape[:2]
M = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]])
sheared_img = cv2.warpPerspective(img2, M, (int(cols*1.5), int(rows*1.5)))
plt.imshow(sheared_img)
```

```python
im1= cv2.imread('/content/back.jpg')
im2= cv2.imread('/content/plant.jfif')
```

```python
im2_resized = cv2.resize(im2, (im1.shape[1], im1.shape[0]))

added_images = cv2.add(im1, im2_resized)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
# Display images
axes[0].imshow(im1)

axes[1].imshow(im2)

axes[2].imshow(added_images)

plt.show()
```

```python
subtracted_images = cv2.subtract(im1, im2_resized)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
# Display images
axes[0].imshow(im1)

axes[1].imshow(im2)

axes[2].imshow(subtracted_images)

plt.show()
```

```python
multiply_images = cv2.multiply(im1, im2_resized)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
# Display images
axes[0].imshow(im1)

axes[1].imshow(im2)

axes[2].imshow(multiply_images)

plt.show()
```

```python
chessboard = np.zeros((8, 8))

# Iterate through the array and alternate the values to represent the chessboard pattern
for i in range(8):
    for j in range(8):
        if (i + j) % 2 == 0:
            chessboard[i, j] = 1

# Plot the chessboard pattern using matplotlib
plt.imshow(chessboard, cmap='binary')
plt.show()
```

```python
chessboard = np.zeros((2, 2))
for i in range(2):
    for j in range(2):
        if (i + j) % 2 == 0:
            chessboard[i, j] = 1

plt.imshow(chessboard, cmap='binary',aspect ='auto')
plt.show()
```

```python
width = 400
height = 400
# Create a black canvas
canvas = np.zeros((height, width), dtype=np.uint8)

# Draw the diagonal line (dividing line)
cv2.line(canvas, (0, 0), (width, height), (255, 255, 255), 2)

# Define the coordinates for the two triangles
pts_black = np.array([[0, 0], [0, height], [width, height]], np.int32)
pts_white = np.array([[0, 0], [width, 0], [width, height]], np.int32)

# Fill the triangles with respective colors
cv2.fillPoly(canvas, [pts_black], (0, 0, 0))
cv2.fillPoly(canvas, [pts_white], (255, 255, 255))

plt.imshow(canvas, cmap='binary')
```

## **DISPLAY DISTANCE PATTERN**

```python
width, height = 400, 400

#blank space
image = np.zeros((height, width), dtype=np.uint8)

center = (width // 2, height // 2)

#Iterate through pixel
for y in range(height):
    for x in range(width):
        distance = np.sqrt((x - center[0])**2 + (y - center[1])**2)

        image[y, x] = 255 if distance < 100 else 0

# Display the distance pattern
cv2_imshow( image)
```
