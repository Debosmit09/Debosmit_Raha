## **EXPERIMENT 3 - Texture feature extractions using GLCM**



**CONCLUSION - In conclusion, GLCM (Gray-Level Co-occurrence Matrix) feature extraction is a powerful technique for texture analysis in images. It characterizes the spatial relationships of pixel intensities within an image. GLCM features provide valuable information regarding texture patterns, such as contrast, correlation, energy, and homogeneity. These features serve as discriminative descriptors for various image classification and segmentation tasks. However, careful selection of parameters and normalization techniques is crucial to enhance the robustness and effectiveness of GLCM-based feature extraction methods.**

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
from skimage.feature import greycomatrix,greycoprops
```

```python
img1=cv2.imread('/content/random.jpeg')
img2=cv2.imread('/content/coarse.jpg')
img3=cv2.imread('/content/smooth.jpg')
img4=cv2.imread('/content/darkbark.jpg')
img5=cv2.imread('/content/bark.jpg')
img6=cv2.imread('/content/pebble.jpg')
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2GRAY)
img5 = cv2.cvtColor(img5, cv2.COLOR_BGR2GRAY)
img6 = cv2.cvtColor(img6, cv2.COLOR_BGR2GRAY)
```

```python
img1 = cv2.resize(img1,(300,300),interpolation=cv2.INTER_LINEAR)
img2 = cv2.resize(img2,(300,300),interpolation=cv2.INTER_LINEAR)
img2 = cv2.resize(img2,(300,300),interpolation=cv2.INTER_LINEAR)
img4 = cv2.resize(img4,(300,300),interpolation=cv2.INTER_LINEAR)
img5 = cv2.resize(img5,(300,300),interpolation=cv2.INTER_LINEAR)
img6 = cv2.resize(img6,(300,300),interpolation=cv2.INTER_LINEAR)
cv2_imshow(img2)
```

## **Gaussian** **filter**

```python
img1 = cv2.GaussianBlur(img1,(5,5),0)
img2 = cv2.GaussianBlur(img2,(5,5),0)
img3 = cv2.GaussianBlur(img3,(5,5),0)
img4 = cv2.GaussianBlur(img4,(5,5),0)
img5 = cv2.GaussianBlur(img5,(5,5),0)
img6 = cv2.GaussianBlur(img6,(5,5),0)
cv2_imshow(img2)
```

```python
distances = [5]  # Define distances for GLCM
angles = [0]  # Define angles (directions) for GLCM
```

```python
glcm_img1=greycomatrix(img1, distances, angles, levels=256, symmetric=True, normed=True)
glcm_img2=greycomatrix(img2, distances, angles, levels=256, symmetric=True, normed=True)
glcm_img3=greycomatrix(img3, distances, angles, levels=256, symmetric=True, normed=True)
glcm_img4=greycomatrix(img4, distances, angles, levels=256, symmetric=True, normed=True)
glcm_img5=greycomatrix(img5, distances, angles, levels=256, symmetric=True, normed=True)
glcm_img6=greycomatrix(img6, distances, angles, levels=256, symmetric=True, normed=True)
```

```python
plt.figure(figsize=(20, 16))

plt.subplot(1, 6, 1)
plt.imshow(img1)

plt.subplot(1, 6, 2)
plt.imshow(img2)

plt.subplot(1, 6, 3)
plt.imshow(img3)

plt.subplot(1, 6, 4)
plt.imshow(img4)

plt.subplot(1, 6, 5)
plt.imshow(img5)

plt.subplot(1, 6, 6)
plt.imshow(img6)
```

```python
properties = ['contrast', 'dissimilarity', 'homogeneity','energy', 'correlation']
images =[glcm_img1,glcm_img2,glcm_img3,glcm_img4,glcm_img5,glcm_img6]
for image in images:
  for prop in properties:
      glcm_properties = graycoprops(image, prop)
      print(prop,':',glcm_properties)
  print("\n")
```

## **If homogeneity in more the texture is smooth .**



```python
# Create a scatter plot for each GLCM property
for prop in properties:
    # Create a list to store property values for each image
    prop_values = []
    # Iterate over each image
    for image in images:
        # Compute GLCM properties for the current image
        glcm_properties = greycoprops(image, prop)
        # Append the property value to the list
        prop_values.append(glcm_properties)
    # Create a scatter plot for the current property
    plt.figure(figsize=(4, 4))
    plt.scatter(range(1, len(images) + 1), prop_values)
    plt.xlabel('Image Index')
    plt.ylabel(prop.capitalize())  # Use property name as y-axis label
    plt.title(f'Comparison of {prop.capitalize()} Across Images')
    plt.xticks(range(1, len(images) + 1))
    plt.grid(True)
    plt.show()
```
